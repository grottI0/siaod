### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
def generate_matr(m=None, n=None, min_limit=None, max_limit=None):
  if m is None and n is None and min_limit is None and max_limit is None:
    m = int(input())
    n = int(input())
    min_limit = float(input())
    max_limit = float(input())

  return [[random.uniform(min_limit, max_limit) for i in range(n)] for j in range(m)]

matr = generate_matr()
for line in matr:

  print(line)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с

временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
start_time = time.time()
def inbuilt(array):
  return array.sort()
matr0 = copy.deepcopy(matr)
for line in matr0:

  print(line)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
 
```

```python
start_time = time.time()

matr1 = copy.deepcopy(matr)

def selection_sort(line):
  for i in range(0, len(line)-1):
    min_index = i
    for j in range(i+1, len(line)):
      if line[j] < line[min_index]:
        min_index = j
      
    if min_index != i:
      line[i], line[min_index] = line[min_index], line[i]
  
  return line

for line in matr1:

  print(selection_sort(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()

matr2 = copy.deepcopy(matr)

def insertion_sort(line):
  for i in range(1, len(line)):
    elem = line[i]
    j = i-1
    while j >= 0 and line[j] > elem:
      line[j+1] = line[j]
      j -= 1
    line[j+1] = elem

  return line
  
for line in matr2:

  print(insertion_sort(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()

matr3 = copy.deepcopy(matr)

def bubble_sort(line):
  while True:
    swaps = 0

    for i in range(len(line) - 1):
      if line[i] > line[i+1]:
        line[i], line[i+1] = line[i+1], line[i]
        swaps += 1

    if swaps == 0:
      break
  
  return line

for line in matr3:
  print(bubble_sort(line))
  
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()

matr4 = copy.deepcopy(matr)

def shell_sort(line):
  step = len(line) // 2
  while step > 0:
    for i in range(step, len(line)):
      elem = line[i]
      j = i
      while j >= step and line[j - step] > elem:
        line[j] = line[j - step]
        j -= step
      line[j] = elem
    
    step //= 2

  return line

for line in matr:
  print(shell_sort(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
matr5 = copy.deepcopy(matr)

def quick_sort(array):
  less = []
  equal = []
  greater = []

  if len(array) > 1:
    pivot = array[0]
    for x in array:
      if x < pivot:
        less.append(x)
      elif x == pivot:
        equal.append(x)
      elif x > pivot:
        greater.append(x)
        
    return quick_sort(less) + equal + quick_sort(greater)
  else:
    return array

for line in matr5:
  print(quick_sort(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.

matr6 = copy.deepcopy(matr)

def heap(array, index, size):
  largest = index
  left = (2*index) + 1
  right = (2*index) + 2

  if left < size and array[left] > array[largest]:
    largest = left
  if right < size and array[right] > array[largest]:
    largest = right
  if largest != index:
    array[largest], array[index] = array[index], array[largest]
    heap(array, largest, size)

def tour_sort(array):
  n = len(array)
  for i in range(n, -1, -1):
    heap(array, 0, n)
  for i in range(n-1, 0, -1):
    array[i], array[0] = array[0], array[i]
    heap(array, 0, i)
  
  return array

start_time = time.time()
for line in matr6:
  print(tour_sort(line))
    

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
import matplotlib.pyplot as plt

def test(sort_method):
  elements = list()
  times = list()
  for i in range(1, 201):
  
      matr7 = generate_matr(m=i ,n=i , min_limit=1, max_limit=20)
      start = time.time()
      for line in matr7:
        sort_method(line)
      end = time.time()
      elements.append(len(matr7[0]) * len(matr7))
      times.append(end-start)
  
  plt.xlabel('Matrix Elements')
  plt.ylabel('Time Complexity')
  plt.plot(elements, times, label = sort_method.__name__)
  plt.grid()
  plt.legend()
  plt.show()

for func in (inbuilt, selection_sort, insertion_sort, bubble_sort, 
             quick_sort, shell_sort, tour_sort):
  test(func)
```
